## 1. Data Filtering

```python
boards = pd.read_excel('board2.xlsx')
boards
```

##### Filename: Instru = ["Name", "GEAS", "Electronics > 70"]; where track is constant as Instrumentation and hometown Luzon

```python
# This code filters a Pandas Dataframe called boards then select columns from the given condition for rows
# .loc is a function accessing a group of rows and columns
Instru = boards.loc[

# the & operator combines all the given condition and show only rows that meets the condition
# this check the Hometown column and selects the row with the value equating to Luzon
    (boards['Hometown'] == 'Luzon') & 
# this check the Track column and selects the row with the value equating to Instrumentation
    (boards['Track'] == 'Instrumentation') & 
# this check the Electronics column and selects the row with the value greater than 70
    (boards['Electronics'] > 70), 
# after the rows have been filtered based on the conditions, it selects only these specified columns
    ['Name','GEAS','Electronics'] ]

Instru
```

##### Filename: Mindy = ["Name", "Track", "Average >= 55"]; where hometown is constant as Mindanao and gender Female


```python
# this select four columns namely Math, Electronics, GEAS, and Communication from the boards dataframe
# it also calculates the average for each individual and rounding it off to two decimal places
# the result is stored in a new column Average.

boards['Average'] = round(boards[['Math', 'Electronics', 'GEAS', 'Communication']].mean(axis=1),2)

# This code filters a Pandas Dataframe called boards then select columns from the given condition for rows
# .loc is a function accessing a group of rows and columns
Mindy = boards.loc[

# the & operator combines all the given condition and show only rows that meets the condition
# this check the Hometown column and selects the row with the value equal to Mindanao
    (boards['Hometown'] == 'Mindanao') & 
# this check the Gender column and selects the row with the value Female
    (boards['Gender'] == 'Female') & 
# this check the Electronics column and selects the row with the value greater than or equal to 55
    (boards['Average'] >= 55), 
# after the rows have been filtered based on the conditions, it selects only these specified columns
    ['Name','Track','Electronics','Average'] ]

Mindy
```

## 2. Data Visualization

##### Create a visualization that shows how the different features contributes to average grade. Does chosen track in college, gender, or hometown, contributes to a higher average score?

```python
# Defines the color palette
sns.set_palette("husl")
# Sets the figure size for the overall plot
plt.figure(figsize=(10,4))
# this creates a bar chart. The values under hometown is used for the x-axis, while the average for the y-axis
plt.bar(boards['Hometown'],boards['Average'])
# this creates a title for the graph, giving some aesthetic to the ouput
plt.title('Average Score by Hometown')


# Sets the figure size for the overall plot
plt.figure(figsize=(10,4))
# this creates a bar chart. The values under track is used for the x-axis, while the average for the y-axis
plt.bar(boards['Track'],boards['Average'])
# this creates a title for the graph, giving some aesthetic to the ouput
plt.title('Average Score by Track')


# Sets the figure size for the overall plot
plt.figure(figsize=(10,4))
# this creates a bar chart. The values under gender is used for the x-axis, while the average for the y-axis
plt.bar(boards['Gender'],boards['Average'])
# this creates a title for the graph, giving some aesthetic to the ouput
plt.title('Average Score by Gender')
```
